레포트 유의 사항

1. 문제의 답안을 함수로 작성하되, 반드시 함수 이름은 문제에서 제시하는 함수명으로 할것 (함수명이 다르면 점수 획득 불가능)
1. 각 문제의 답안은 답안 작성 셀에서 작성할 것
1. 모든 답안에 주석을 달지 말것(주석을 달면 0점 처리)
1. 모든 문제는 함수의 매개변수 입력을 통해 데이터를 전달 받고, 결과를 반환하는 함수를 정의하는 것임 (주의: input()는 키보드 입력이며, 함수 입력이 아님, print()는 모니터에 출력하는 함수이며, 함수 반환이 아님)
1. 함수의 입력과 반환 시 데이터 타입도 반드시 지킬것 (예: 문자열 반환이라는 문제에서는 반드시 문자열을 반환 해야함)
1. 답안 제출은 함수가 정의된 부분만 제출하고, 나머지 코드는 지우고 제출할 것, 답안 제출시 input() 함수나 , print() 함수가 있으면 안됨
1. 코드 작성이 끝난 파일은 개별적으로 함수를 호출해서 검수를 거쳐 정답이 맞는지 확인할 것
1. 작성된 답안은 colab의 상단 메뉴의 파일--> 다운로드 --> .py 로 다운 받아 학교 레포트실에 제출할 것  ( .py 다운로드시 경고창이 뜨지만, 계속 버튼을 눌러 다운 할것)
1. .py 가 레포트실에 올라가지 않을경우 ZIP으로 압축하여 업로드 가능
1. 파일의 이름은 자신의 학번.py      예) 학번이 2021111000 일 경우,     2021111000.py 파일을 제출
1. 코드는 자동 채점이 되며, 각 문제당 부분 점수는 없음 (문제에 추가된 외국어는 무시할것)
1. 시험 문제 및 답안을 절대 유출 금지(SNS, 커뮤니티 및 개인 블로그 포함)하며, 유출될 경우 불이익을 감수할 것
1. AI로 문제를 푸는 것을 금지하며, AI로 푼 답안을 제출할 경우 0점 처리(타 학생과 유사한 답안 작성도 0점 처리)
1. 제출 기한을 넘겨 답안을 제출하는 것은 불가

## 문제 1) - 5점

![q1](https://github.com/ykkim77/final-report/blob/main/q1.png?raw=1)

```python
###### 1번 문제 답안 작성 셀 ######
def exam1(a, b, d):

    scores = [a, b, d]
    for score in scores:

        if not (0 <= score <= 100) or (score % 2 != 0):
            return "Invalid Value"

    avg = (a + b + d) / 3


    if 90 <= avg <= 100:
        grade = "A"
    elif 80 <= avg < 90:
        grade = "B"
    elif 70 <= avg < 80:
        grade = "C"
    elif 60 <= avg < 70:
        grade = "D"
    else:
        grade = "F"


    return grade
```

## 문제 7) - 15점

![q7](https://github.com/ykkim77/final-report/blob/main/q7.png?raw=1)

```python
###### 7번 문제 답안 작성 셀 ######
def exam7(a):
    words = a.split()
    result_words = []
    for word in words:
        reversed_word = word[::-1]
        if len(reversed_word) % 2 == 0:
            result_words.append(reversed_word.upper())
        else:
            result_words.append(reversed_word)
    b = " ".join(result_words)
    return b

```

## 문제 9) - 15점

![q9](https://github.com/ykkim77/final-report/blob/main/q9.png?raw=1)

```python
###### 9번 문제 답안 작성 셀 ######
def exam9(a):
    counts = {}
    order = []
    for char in a:
        if char not in counts:
            counts[char] = 1
            order.append(char)
        else:
            counts[char] += 1
    sorted_order = sorted(order, key=lambda x: counts[x], reverse=True)
    b = [{sorted_order[0]: counts[sorted_order[0]]}, {sorted_order[1]: counts[sorted_order[1]]}]
    return b

```

## 문제 10) - 15점

![q10](https://github.com/ykkim77/final-report/blob/main/q10.png?raw=1)

```python
###### 10번 문제 답안 작성 셀######

def exam10(a):
    counts = {}
    for group in a:
        for member in group:
            if member not in counts:
                counts[member] = 1
            else:
                counts[member] += 1
    max_key = ""
    max_val = 0
    for key in counts:
        if counts[key] > max_val:
            max_val = counts[key]
            max_key = key
    b = {max_key: max_val}
    return b
```
